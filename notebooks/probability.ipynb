# Probability


### Important Terms

- **Sample Space**: The set of all possible outcomes of an experiment.
- **Event**: A subset of the sample space, which can include one or more outcomes.
- **Outcome**: A single result of an experiment.

let's say we flip a coin 10 times. The sample space would be all the possible sequences of heads (H) and tails (T) that can occur in 10 flips. An event could be getting exactly 5 heads and 5 tails, which is a specific subset of the sample space. what's the probability of getting exactly 5 heads in 10 flips?
$$P(\text{exactly 5 heads in 10 flips}) = \frac{\text{Number of favorable outcomes}}{\text{Total number of outcomes}}= \frac{5}{10}= 0.5$$

This example is **Theoretical Probability**, which is based on the assumption that all outcomes are equally likely, but in practice, we often deal with **Empirical Probability**, which is based on observed data.

To make Empirical Probability like Theoretical Probability, we can conduct a large number of trials and observe the outcomes. The more trials we conduct, the closer our empirical probability will be to the theoretical probability. $\lim\limits_{N \to \infty} P(A) = \frac{N(A)}{N}$

```python
import random
import matplotlib.pyplot as plt

def EmpiricalProbForCoin(n):
    head_count = 0
    tail_count = 0
    head_probs = []
    tail_probs = []

    for i in range(1, n + 1):
        if random.randint(0, 1) == 0:
            head_count += 1
        else:
            tail_count += 1

        head_probs.append(head_count / i)
        tail_probs.append(tail_count / i)

    # Final results
    print("Heads:", head_count, "Tails:", tail_count)
    print("Empirical Probability of Heads:", head_count / n)
    print("Empirical Probability of Tails:", tail_count / n)

    # Plotting
    plt.figure(figsize=(10, 6))

    plt.plot(range(1, n + 1), head_probs, label='Empirical P(Heads)', color='blue')
    plt.plot(range(1, n + 1), tail_probs, label='Empirical P(Tails)', color='orange')
    plt.axhline(y=0.5, color='gray', linestyle='--', label='Theoretical P = 0.5')

    plt.xlabel('Number of Trials')
    plt.ylabel('Probability')
    plt.title('Empirical Probability of Coin Flips Over Time')
    plt.grid(True)
    plt.legend()
    plt.show()
```

#### Example: Empirical Probability of Coin Flips 10 times

```python
EmpiricalProbForCoin(10)
```

##### Example: Empirical Probability of Coin Flips 1000 times

```python
EmpiricalProbForCoin(1000)
```

##### Example: Empirical Probability of Coin Flips 100000 times

```python
EmpiricalProbForCoin(100000)
```

### Random Variables

A **Random Variable** is a variable whose value is subject to chance. It can take on different values based on the outcome of a random event. Random variables can be classified into two types:

- **Discrete Random Variables**: These can take on a countable number of values. For example, the number of heads in 10 coin flips.
- **Continuous Random Variables**: These can take on an infinite number of values within a given range. For example, the height of a person.

As an example let's say we want $P(X = \text{number of tails})$. we have $(H,H),(H,T),(T,H),(T,T)$, so $P(X = 0) = \frac{1}{4}, P(X = 1) = \frac{2}{4}, P(X = 2) = \frac{1}{4}$


```python
sampleSpace = ['HH', 'HT', 'TH', 'TT']

for outcome in sampleSpace:
    tails = outcome.count('T')
    print(f"P(X = {tails}) = {sampleSpace.count(outcome)}/{len(sampleSpace)} = {sampleSpace.count(outcome)/len(sampleSpace)}")

plt.figure(figsize=(10, 6))
plt.bar(['0 Tails', '1 Tail', '2 Tails'], [1/4, 2/4, 1/4], color=['blue', 'orange', 'green'])
plt.xlabel('Number of Tails')
plt.ylabel('Probability')
plt.title('Probability Distribution of Tails in Coin Flips')

plt.grid(True)
plt.show()
```

### Multivariate Probability

- **Joint Probability**: The probability of two or more events happening at the same time. For example, the probability of getting heads on the first flip and tails on the second flip.
- **Marginal Probability**: The probability of a single event occurring, regardless of the outcome of other events. For example, the probability of getting heads on the first flip, regardless of the second flip.
- **Conditional Probability**: The probability of an event occurring given that another event has already occurred. For example, the probability of getting heads on the second flip given that the first flip was heads.

as an example let's say we have this table:
|  | Pass | Fail | Total |
|---|------|------|------|
|Male| 30 | 10 | 40 |
|Female| 20 | 20 | 40 |
|Total| 50 | 30 | 80 |


```python
data = {'male':[30, 10], 'female':[20, 20]}

print(f"Joint Probability: P(pass & male) = {30/40}")
print(f"Marginal Probability: P(pass) = {50/80}")
```

Finally, important thing the **mutually exclusive events**. Two events are mutually exclusive if they cannot occur at the same time. For example, getting heads and tails on the same coin flip are mutually exclusive events. The probability of either event occurring is the sum of their individual probabilities.
